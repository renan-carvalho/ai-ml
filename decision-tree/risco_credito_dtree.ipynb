```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open ('/content/drive/MyDrive/Colab Notebooks/risco_credito.pkl', 'rb') as f:
  X_credit_train, y_credit_train = pickle.load(f)
```

```python
X_credit_train.shape, y_credit_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_credit_model = DecisionTreeClassifier(criterion='entropy', random_state=0)
tree_credit_model.fit(X_credit_train, y_credit_train)
```

```python
previsoes = tree_credit_model.predict([[0,0,1,2],[2,0,0,0]])
previsoes
```

```python
from sklearn import tree
previsores = ['historia', 'divida', 'garantias', 'renda']
figura, eixos = plt.subplots(nrows=1,ncols=1, figsize=(10,10))
tree.plot_tree(tree_credit_model, feature_names=previsores, class_names=tree_credit_model.classes_, filled=True);
```
