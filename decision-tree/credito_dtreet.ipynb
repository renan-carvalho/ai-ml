```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open ('/content/drive/MyDrive/Colab Notebooks/credit.pkl', 'rb') as f:
  X_credit_train, y_credit_train, X_credit_test, y_credit_test = pickle.load(f)
```

```python
X_credit_train.shape, y_credit_train.shape
```

```python
X_credit_test.shape, y_credit_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_credit_model = DecisionTreeClassifier(criterion='entropy', random_state=0)
tree_credit_model.fit(X_credit_train, y_credit_train)
```

```python
previsoes = tree_credit_model.predict(X_credit_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_credit_test, previsoes)
```

```python
confusion_matrix(y_credit_test, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(tree_credit_model)
cm.fit(X_credit_train, y_credit_train)
cm.score(X_credit_test, y_credit_test)
```

```python
print(classification_report(y_credit_test, previsoes))
```

```python
from sklearn import tree
previsores = ['person_income','person_age','loan_amnt']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (20,20))
tree.plot_tree(tree_credit_model, feature_names=previsores, class_names=str(tree_credit_model.classes_), filled=True);
fig.savefig('arvore_credit.png')
```
